```python
# Average age of individuals
avg_age = df['Age'].mean()
print("Average Age:", avg_age)
```

```python
# Occupation with highest average physical activity level
occupation_avg_activity = df.groupby('Occupation')['Physical Activity Level'].mean()
highest_activity_occupation = occupation_avg_activity.idxmax()
print("Occupation with Highest Average Activity:", highest_activity_occupation)
```

```python
# Top 5 individuals with highest quality of sleep
top_quality_sleep = df.nlargest(5, 'Quality of Sleep')
print("Top 5 Quality of Sleep:\n", top_quality_sleep)
```

```python
# Individuals with sleep disorder and are doctors
sleep_disorder_doctors = df[(df['Sleep Disorder'] == 'Sleep Apnea') & (df['Occupation'] == 'Doctor')]
print("Doctors with Sleep Disorder:\n", sleep_disorder_doctors)
```

```python
# Correlation between sleep duration and stress level
correlation = df['Sleep Duration'].corr(df['Stress Level'])
print("Correlation between Sleep Duration and Stress Level:", correlation)
```

```python
# Average heart rate for males and females
avg_heart_rate_gender = df.groupby('Gender')['Heart Rate'].mean()
print("Average Heart Rate by Gender:\n", avg_heart_rate_gender)
```

### Piechart
- plt.pie()

```python
category_counts = df['Sleep Disorder'].value_counts()

plt.figure(figsize=(8, 8))  # Set the figure size (optional)

# Plot the pie chart
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140,shadow=True)

# Add a title to the pie chart
plt.title('Distribution of Sleep Disorder')

# Show the pie chart
plt.show()
```

### Categorical distribution data
- Boxplot

```python
plt.figure(figsize=(10, 6))  # Set the figure size (optional)
sns.boxplot(x='Sleep Disorder', y='Sleep Duration', data=df)

# Add labels and title to the plot
plt.xlabel('X-axis Label (Categorical Column)')
plt.ylabel('Y-axis Label (Numerical Column)')
plt.title('Box Plot - Categorical Distribution Data')

# Show the boxplot
plt.show()
```

```python
sns.boxplot(x=df['Stress Level'])
```

### violin plot
- Similar to boxplot

```python
plt.figure(figsize=(10, 6))  # Set the figure size (optional)

sns.violinplot(x='Quality of Sleep', y='Heart Rate', data=df)

# Add labels and title to the plot
plt.xlabel('X-axis Label (Categorical Column)')
plt.ylabel('Y-axis Label (Numerical Column)')
plt.title('Violin Plot - Categorical Distribution Data')

# Show the violin plot
plt.show()
```

```python
sns.violinplot(x=df['Sleep Duration'])
```
